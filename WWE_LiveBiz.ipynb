```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,6)
```

```python
wwe_events = pd.DataFrame({
    "Region": ["USA", "USA", "Europe", "Europe"],
    "Event_Count": [220, 180, 95, 70],
    "Total_Revenue_MUSD": [520, 430, 210, 160],
    "Operating_Cost_MUSD": [360, 300, 135, 110],
    "Cash_Flow_MUSD": [160, 130, 75, 50],
    "Current_Assets_MUSD": [900, 900, 450, 450],
    "Current_Liabilities_MUSD": [500, 500, 250, 250],
    "Recurring_Revenue_%": [68, 68, 60, 60]
})

wwe_events
```

```python
wwe_events["Net_Profit_MUSD"] = (
    wwe_events["Total_Revenue_MUSD"] - wwe_events["Operating_Cost_MUSD"]
)

wwe_events["Profit_Margin_%"] = (
    wwe_events["Net_Profit_MUSD"] / wwe_events["Total_Revenue_MUSD"] * 100
)

wwe_events["Current_Ratio"] = (
    wwe_events["Current_Assets_MUSD"] / wwe_events["Current_Liabilities_MUSD"]
)

wwe_events[[
    "Region",
    "Net_Profit_MUSD",
    "Profit_Margin_%",
    "Cash_Flow_MUSD",
    "Current_Ratio",
    "Recurring_Revenue_%"
]]
```

```python
plt.bar(wwe_events["Region"], wwe_events["Profit_Margin_%"])
plt.title("WWE Live Events Profit Margins (USA vs Europe)")
plt.ylabel("Profit Margin (%)")
plt.xlabel("Region")
plt.show()
```
